# Lemmatization

## What is Lemmatization?

**Lemmatization** is a text preprocessing technique in Natural Language Processing (NLP) that converts a word into its **base or dictionary form**, known as the **lemma**.

Unlike stemming, lemmatization considers the **meaning (context)** and **part of speech (POS)** of a word before converting it to its root form. As a result, the output is always a meaningful dictionary word.

---

## Example

| Original Word | Lemma |
|---------------|-------|
| running | run |
| ran | run |
| studies | study |
| studying | study |
| better | good |
| cars | car |
| leaves | leaf |
| mice | mouse |

> **Note:** The lemma is always a valid dictionary word.

---

## Stemming vs Lemmatization

| Stemming | Lemmatization |
|-----------|---------------|
| Removes prefixes or suffixes. | Uses vocabulary and linguistic rules. |
| May produce invalid words. | Produces valid dictionary words. |
| Faster. | Slightly slower but more accurate. |
| Ignores word meaning. | Considers word meaning and Part of Speech (POS). |

Example:

```
Stemming:
studies → studi

Lemmatization:
studies → study
```

# WordNet Lemmatizer

## What is WordNet?

**WordNet** is a large lexical database of the English language developed at Princeton University. It groups English words into sets of synonyms (called **synsets**) and stores relationships between words.

The **WordNet Lemmatizer** uses this dictionary to convert words into their correct base (lemma) form.

---

## How WordNet Lemmatizer Works

1. Takes the input word.
2. Identifies its Part of Speech (POS), if provided.
3. Searches the WordNet dictionary.
4. Returns the correct dictionary form (lemma).

---

## Examples

| Input Word | POS | Output |
|------------|-----|--------|
| running | Verb | run |
| studies | Verb | study |
| better | Adjective | good |
| cars | Noun | car |
| wolves | Noun | wolf |

---

## Advantages of WordNet Lemmatizer

- Produces meaningful dictionary words.
- More accurate than stemming.
- Considers the context and Part of Speech (POS).
- Widely used in NLP applications such as sentiment analysis, chatbots, and search engines.

---

## Disadvantages

- Slower than stemming.
- Requires a lexical database (WordNet).
- Results are best when the correct POS tag is provided.

---

## Summary

- **Lemmatization** converts words into their dictionary (base) form.
- **WordNet Lemmatizer** uses the WordNet lexical database to find the correct lemma.
- Unlike stemming, lemmatization produces meaningful English words.
- It is more accurate but computationally more expensive than stemming.
