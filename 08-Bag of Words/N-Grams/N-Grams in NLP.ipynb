# N-Grams in NLP

## Definition

**N-Grams** are contiguous sequences of **N words (or tokens)** that appear together in a sentence or document.

The value of **N** determines the number of words in each sequence.

- **N = 1** → Unigram
- **N = 2** → Bigram
- **N = 3** → Trigram
- **N = 4** → Four-gram
- **N = 5** → Five-gram

> **Definition:**  
> An **N-Gram** is a sequence of **N consecutive words** extracted from a sentence while preserving the word order.

---

# Why Do We Use N-Grams?

Unlike the **Bag of Words (BoW)** model, which ignores word order, **N-Grams preserve the local context** by considering neighboring words.

For example:

```text
I love Machine Learning
```

Bag of Words treats each word independently, whereas N-Grams capture relationships like:

- I love
- love Machine
- Machine Learning

This helps models understand phrases instead of isolated words.

---

# Example Sentence

```text
I love Natural Language Processing
```

---

## 1. Unigram (N = 1)

Each individual word is treated as a separate token.

```text
I

love

Natural

Language

Processing
```

Output:

```python
['I', 'love', 'Natural', 'Language', 'Processing']
```

---

## 2. Bigram (N = 2)

Every two consecutive words form a pair.

```text
I love

love Natural

Natural Language

Language Processing
```

Output:

```python
[
 ('I', 'love'),
 ('love', 'Natural'),
 ('Natural', 'Language'),
 ('Language', 'Processing')
]
```

---

## 3. Trigram (N = 3)

Every three consecutive words form a sequence.

```text
I love Natural

love Natural Language

Natural Language Processing
```

Output:

```python
[
 ('I', 'love', 'Natural'),
 ('love', 'Natural', 'Language'),
 ('Natural', 'Language', 'Processing')
]
```

---

## 4. Four-Gram (N = 4)

```text
I love Natural Language

love Natural Language Processing
```

Output:

```python
[
 ('I', 'love', 'Natural', 'Language'),
 ('love', 'Natural', 'Language', 'Processing')
]
```

# Formula

If a sentence contains **W words**, then:

```text
Number of N-Grams = W − N + 1
```

### Example

Sentence:

```text
I love NLP
```

Number of words:

```text
W = 3
```

For **Bigram (N = 2):**

```text
3 − 2 + 1 = 2
```

Bigrams:

```text
I love

love NLP
```

---

# Python Implementation using NLTK

```python
from nltk.util import ngrams

sentence = "I love Natural Language Processing"

words = sentence.split()

# Bigrams
bigrams = list(ngrams(words, 2))
print("Bigrams:")
print(bigrams)

# Trigrams
trigrams = list(ngrams(words, 3))
print("\nTrigrams:")
print(trigrams)
```

### Output

```python
Bigrams:
[
 ('I', 'love'),
 ('love', 'Natural'),
 ('Natural', 'Language'),
 ('Language', 'Processing')
]

Trigrams:
[
 ('I', 'love', 'Natural'),
 ('love', 'Natural', 'Language'),
 ('Natural', 'Language', 'Processing')
]
```

---

# Python Implementation using Scikit-learn

```python
from sklearn.feature_extraction.text import CountVectorizer

corpus = [
    "I love NLP",
    "NLP is amazing"
]

# Create Bigram Model
vectorizer = CountVectorizer(ngram_range=(2, 2))

X = vectorizer.fit_transform(corpus)

print(vectorizer.get_feature_names_out())
print(X.toarray())
```

### Output

```python
['is amazing' 'love nlp' 'nlp is']

[[0 1 0]
 [1 0 1]]
```

# Advantages

- ✅ Preserves word order.
- ✅ Captures local context.
- ✅ Improves text classification performance.
- ✅ Detects common phrases.
- ✅ Better than Bag of Words for phrase-based analysis.

---

# Disadvantages

- ❌ Vocabulary size grows rapidly as **N** increases.
- ❌ Produces sparse feature vectors.
- ❌ Requires more memory and computation.
- ❌ Cannot capture long-range dependencies.
- ❌ Larger N-Grams may not generalize well on small datasets.

---

# Bag of Words vs N-Grams

| Feature | Bag of Words | N-Grams |
|----------|--------------|---------|
| Considers Word Order | ❌ No | ✅ Yes |
| Captures Context | ❌ No | ✅ Partially |
| Uses Word Frequency | ✅ Yes | ✅ Yes |
| Captures Phrases | ❌ No | ✅ Yes |
| Vocabulary Size | Moderate | Larger |
| Feature Sparsity | High | Very High |

---

# Summary

**N-Grams** are sequences of **N consecutive words** extracted from text while preserving their order. Unlike the **Bag of Words** model, N-Grams capture local context and common phrases, making them more effective for tasks such as text classification, language modeling, and next-word prediction. However, as the value of **N** increases, the vocabulary size and computational cost also increase.
