# Bag of Words (BoW) in NLP

## Definition

**Bag of Words (BoW)** is a text representation technique in **Natural Language Processing (NLP)** that converts text into numerical vectors based on the **frequency of words** in a document.

- It ignores the **order** of words.
- It considers only the **occurrence (presence)** or **frequency** of words.
- It is one of the simplest and most widely used feature extraction techniques in NLP.

> **Formula:**
>
> If the vocabulary contains **V** unique words, each document is represented by a vector of length **V**, where each value represents the count of the corresponding word.

---

# How Bag of Words Works

The Bag of Words model follows these steps:

1. Collect all documents.
2. Perform text preprocessing (lowercase, remove punctuation, etc.).
3. Create a vocabulary of unique words.
4. Count the frequency of each word in every document.
5. Represent each document as a numerical vector.

# Example

## Documents

```text
Document 1: I love NLP

Document 2: I love AI

Document 3: AI loves NLP
```

---

## Step 1: Build Vocabulary

```python
Vocabulary = ["AI", "I", "love", "loves", "NLP"]
```

Vocabulary Size = **5**

---

## Step 2: Count Word Frequencies

| Word | Doc 1 | Doc 2 | Doc 3 |
|------|:-----:|:-----:|:-----:|
| AI | 0 | 1 | 1 |
| I | 1 | 1 | 0 |
| love | 1 | 1 | 0 |
| loves | 0 | 0 | 1 |
| NLP | 1 | 0 | 1 |

---

## Step 3: Vector Representation

### Document 1

```text
"I love NLP"

[0, 1, 1, 0, 1]
```

---

### Document 2

```text
"I love AI"

[1, 1, 1, 0, 0]
```

---

### Document 3

```text
"AI loves NLP"

[1, 0, 0, 1, 1]
```

---

# Python Implementation

```python
from sklearn.feature_extraction.text import CountVectorizer

# Sample Documents
documents = [
    "I love NLP",
    "I love AI",
    "AI loves NLP"
]

# Create CountVectorizer
vectorizer = CountVectorizer()

# Fit and Transform
bow_matrix = vectorizer.fit_transform(documents)

# Vocabulary
print("Vocabulary:")
print(vectorizer.get_feature_names_out())

# BoW Matrix
print("\nBag of Words Matrix:")
print(bow_matrix.toarray())
```

---

## Output

```python
Vocabulary:
['ai' 'love' 'loves' 'nlp']

Bag of Words Matrix:

[[0 1 0 1]
 [1 1 0 0]
 [1 0 1 1]]
```

> **Note:** `CountVectorizer` automatically converts text to lowercase by default, so `"I"` is ignored because it is treated as a stop word in many preprocessing settings.

# Advantages

- ✅ Simple and easy to understand.
- ✅ Easy to implement.
- ✅ Converts text into numerical features.
- ✅ Works well for basic text classification tasks.
- ✅ Fast for small datasets.

---

# Disadvantages

- ❌ Ignores word order.
- ❌ Cannot capture context or semantics.
- ❌ Produces sparse and high-dimensional vectors.
- ❌ Vocabulary grows with dataset size.
- ❌ Cannot understand synonyms.

---

# Example Showing the Limitation

Consider these two sentences:

```text
Sentence 1:
Dog bites man

Sentence 2:
Man bites dog
```

Bag of Words Representation:

```text
Dog   Man   Bites

1      1      1

1      1      1
```

Both sentences produce the **same vector**, even though their meanings are completely different.

This is because **Bag of Words ignores the order of words**.

# Bag of Words vs One-Hot Encoding

| Feature | One-Hot Encoding | Bag of Words |
|----------|------------------|--------------|
| Represents | Single Word | Entire Document |
| Uses Frequency | ❌ No | ✅ Yes |
| Vocabulary Required | ✅ Yes | ✅ Yes |
| Sparse Representation | ✅ Yes | ✅ Yes |
| Captures Word Order | ❌ No | ❌ No |
| Captures Meaning | ❌ No | ❌ No |
| Input Type | Word | Sentence/Document |

---

# Summary

Bag of Words (BoW) is a feature extraction technique that represents a document using the frequency of words in a predefined vocabulary. It is simple, fast, and effective for many classical machine learning algorithms. However, it ignores word order and semantic relationships, making it less suitable for modern NLP tasks. Today, techniques such as **TF-IDF**, **Word2Vec**, **GloVe**, **FastText**, and **Transformer-based embeddings (e.g., BERT)** are preferred because they capture contextual and semantic information.
