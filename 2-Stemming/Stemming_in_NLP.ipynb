# Stemming

## What is Stemming?

**Stemming** is a text preprocessing technique in Natural Language Processing (NLP) that reduces a word to its **root form (stem)** by removing prefixes or suffixes.

The resulting stem **may not always be a valid English word**, but it helps normalize text by treating different forms of the same word as a single term.

---

## Example

| Original Word | Stem |
|---------------|------|
| playing | play |
| played | play |
| playing | play |
| walks | walk |
| walking | walk |
| studies | studi |
| studying | studi |
| caring | care |

> **Note:** The word **"studies" → "studi"** is not a valid English word. This is expected because stemming simply removes word endings without considering the word's meaning.

---

## Why Do We Use Stemming?

- Reduces the vocabulary size.
- Converts similar words into a common root.
- Improves the efficiency of Machine Learning models.
- Reduces computation and memory usage.
- Commonly used in text classification, search engines, and spam detection.

---

## Advantages

- Fast and computationally efficient.
- Easy to implement.
- Reduces feature dimensions.
- Works well for Information Retrieval (IR) tasks.

---

## Disadvantages

- The stem may not be a valid dictionary word.
- Can remove too much or too little of a word.
- Less accurate than Lemmatization because it ignores the context and meaning of words.

---

## Types of Stemmers

### 1. Porter Stemmer
- Most popular and widely used stemming algorithm.
- Produces good results for general NLP tasks.
- Balanced between accuracy and speed.

### 2. Regexp (Regex) Stemmer
- Uses **Regular Expressions (Regex)** to remove suffixes based on user-defined patterns.
- Gives more control over how words are stemmed.
- Useful when you want to remove specific endings.

### 3. Snowball Stemmer
- An improved version of the Porter Stemmer.
- Supports multiple languages.
- More efficient and flexible than Porter Stemmer.

### 4. Lancaster Stemmer
- An aggressive stemming algorithm.
- Removes more characters than Porter and Snowball.
- Can sometimes over-stem words.

---

## Summary

| Stemmer | Description |
|----------|-------------|
| Porter Stemmer | Most commonly used and balanced stemmer. |
| Regexp (Regex) Stemmer | Uses regular expressions to remove custom suffixes. |
| Snowball Stemmer | Improved version of Porter with multi-language support. |
| Lancaster Stemmer | Very aggressive stemmer that may over-stem words. |
