# Word2Vec in NLP

## What is Word2Vec?

**Word2Vec** is a **word embedding** technique that converts words into **dense numerical vectors** while preserving their semantic meaning. It learns these vector representations by analyzing the context in which words appear.

Unlike **One-Hot Encoding** and **Bag of Words (BoW)**, Word2Vec captures the semantic relationship between words. As a result, words with similar meanings have similar vector representations.

### Example

```text
King   → [0.25, -0.81, 0.47, ..., 0.15]
Queen  → [0.23, -0.79, 0.49, ..., 0.18]
Man    → [0.12, -0.56, 0.33, ..., 0.09]
Woman  → [0.11, -0.54, 0.35, ..., 0.10]
```

Notice that **King** and **Queen** have similar vector representations because they appear in similar contexts.

---

# Why Do We Need Word2Vec?

Traditional text representation techniques have several limitations.

### One-Hot Encoding

* Produces high-dimensional sparse vectors.
* Does not capture semantic meaning.
* Similar words are treated as completely different.

### Bag of Words (BoW)

* Counts word frequency.
* Ignores word order and context.
* Cannot understand relationships between words.

### Word2Vec Solves These Problems

* Produces dense, low-dimensional vectors.
* Captures semantic meaning.
* Similar words have similar vector representations.
* Improves the performance of many NLP models.

---

# Intuition Behind Word2Vec

Word2Vec is based on the **Distributional Hypothesis**:

> **"Words that appear in similar contexts tend to have similar meanings."**

For example,

```text
The cat is sleeping.
The dog is sleeping.
```

Since **cat** and **dog** appear in almost the same context, Word2Vec learns similar vector representations for both words.

---

# How Word2Vec Works

1. Read a large collection of text (corpus).
2. For each word, observe the surrounding words (context).
3. Train a shallow neural network to learn relationships between words.
4. Store each word as a dense vector (embedding).

After training, every word has its own numerical representation that captures its meaning.

---

# Word2Vec Architectures

Word2Vec has **two training algorithms**:

1. **CBOW (Continuous Bag of Words)**
2. **Skip-Gram**

Both produce word embeddings, but they learn them differently.

---

# 1. CBOW (Continuous Bag of Words)

## Definition

**CBOW predicts the target word using its surrounding context words.**

### Easy Formula

```text
Context Words → Target Word
```

---

## Example

Sentence:

```text
The cat is sitting on the mat.
```

Training Example:

```text
Input  : cat, is, on, the
Output : sitting
```

The model learns to predict the missing word using the surrounding words.

---

## Working of CBOW

### Step 1: Input Context Words

```text
cat
is
on
the
```

↓

### Step 2: Convert each word into its embedding vector.

↓

### Step 3: Average the embedding vectors.

$$
\mathbf{v}=\frac{\mathbf{v_1}+\mathbf{v_2}+\mathbf{v_3}+\cdots+\mathbf{v_n}}{n}
$$

where:

* $\mathbf{v_i}$ = Embedding of the $i^{th}$ context word
* $n$ = Number of context words

↓

### Step 4: Predict the target word using a Softmax layer.

```text
Prediction

↓

sitting
```

---

## Advantages of CBOW

* Fast training
* Computationally efficient
* Works well for frequent words
* Less memory usage

---

## Disadvantages of CBOW

* Performs poorly on rare words
* Ignores word order
* May lose contextual information because it averages embeddings

---

# 2. Skip-Gram

## Definition

**Skip-Gram predicts the surrounding context words using the target word.**

### Easy Formula

```text
Target Word → Context Words
```

---

## Example

Sentence:

```text
The cat is sitting on the mat.
```

Training Example:

```text
Input  : sitting

Output : cat
         is
         on
         the
```

The model learns to predict the surrounding words from the target word.

---

## Working of Skip-Gram

### Step 1: Input the target word.

```text
sitting
```

↓

### Step 2: Convert the target word into an embedding vector.

↓

### Step 3: Predict the surrounding context words.

```text
cat
is
on
the
```

---

## Advantages of Skip-Gram

* Better performance for rare words
* Learns richer word representations
* Usually achieves higher accuracy than CBOW

---

## Disadvantages of Skip-Gram

* Slower training
* Higher computational cost
* Requires more training data

---

# CBOW vs Skip-Gram

| Feature            | CBOW                | Skip-Gram             |
| ------------------ | ------------------- | --------------------- |
| Goal               | Predict target word | Predict context words |
| Input              | Context words       | Target word           |
| Output             | Target word         | Context words         |
| Training Speed     | Fast                | Slow                  |
| Computational Cost | Low                 | High                  |
| Best For           | Frequent words      | Rare words            |

```text
Input:

I
love
learning

↓

Output:

machine
```

# Important Word2Vec Parameters

| Parameter     | Description                                                         |
| ------------- | ------------------------------------------------------------------- |
| `vector_size` | Dimension of the word embedding (e.g., 100, 200, 300)               |
| `window`      | Number of neighboring words considered as context                   |
| `min_count`   | Minimum word frequency required to include a word in the vocabulary |
| `workers`     | Number of CPU cores used during training                            |
| `sg=0`        | Uses the **CBOW** algorithm                                         |
| `sg=1`        | Uses the **Skip-Gram** algorithm                                    |

# Applications of Word2Vec

* Text Classification
* Sentiment Analysis
* Machine Translation
* Information Retrieval
* Search Engines
* Recommendation Systems
* Chatbots
* Document Similarity
* Topic Modeling

---

# Limitations of Word2Vec

* Generates only one embedding per word.
* Cannot distinguish different meanings of the same word (e.g., **bank** as a financial institution vs. a river bank).
* Cannot handle unseen (out-of-vocabulary) words after training.
* Requires a large dataset to learn high-quality embeddings.

---

# Key Takeaways

* **Word2Vec** converts words into dense numerical vectors.
* It learns word meanings from the surrounding context.
* Words with similar meanings have similar vector representations.
* **CBOW** predicts the **target word** from **context words**.
* **Skip-Gram** predicts **context words** from a **target word**.
* **CBOW** is faster and works well for common words.
* **Skip-Gram** is slower but performs better for rare words.
* Word2Vec is widely used in modern NLP applications to create meaningful word embeddings.
