```python
from gensim.models import Word2Vec
```

```python
sentences = [
    ["i", "love", "machine", "learning"],
    ["machine", "learning", "is", "awesome"],
    ["i", "love", "python"],
    ["python", "is", "easy"],
    ["deep", "learning", "uses", "neural", "networks"]
]
```

```python
model = Word2Vec(
    sentences,
    vector_size=100,   # Size of word embeddings
    window=5,          # Context window size
    min_count=1,       # Ignore words with frequency < 1
    workers=4,
    sg=0               # 0 = CBOW, 1 = Skip-Gram
)
```

```python
print(model.wv.index_to_key)
```

```python
vector = model.wv['machine']
```

```python
print(vector)
```

```python
model.wv.most_similar('machine')
```

```python
similarity = model.wv.similarity("machine", "learning")

print(similarity)
```

```python
model.save("word2vec.model")
```

## CBOW vs Skip-Gram in Code

### CBOW (Default)

```python
model = Word2Vec(
    sentences,
    vector_size=100,
    window=5,
    min_count=1,
    sg=0
)
```

### Skip-Gram

```python
model = Word2Vec(
    sentences,
    vector_size=100,
    window=5,
    min_count=1,
    sg=1
)
```

| Parameter     | Description                                          |
| ------------- | ---------------------------------------------------- |
| `sentences`   | Tokenized text data                                  |
| `vector_size` | Size of each word vector (e.g., 100, 200, 300)       |
| `window`      | Number of neighboring words considered as context    |
| `min_count`   | Minimum frequency required for a word to be included |
| `workers`     | Number of CPU cores used for training                |
| `sg=0`        | Uses the CBOW algorithm                              |
| `sg=1`        | Uses the Skip-Gram algorithm                         |
