# TF-IDF (Term Frequency – Inverse Document Frequency)

## What is TF-IDF?

**TF-IDF (Term Frequency – Inverse Document Frequency)** is a statistical technique used in **Natural Language Processing (NLP)** and **Information Retrieval (IR)** to measure how important a word is in a document relative to a collection of documents (corpus).

The main idea behind TF-IDF is:

- Words that occur **frequently in a document** are considered important.
- Words that occur **frequently across many documents** (e.g., *the*, *is*, *and*) are considered less informative.
- TF-IDF assigns **higher weights** to words that are frequent in a specific document but rare across the corpus.

---

# Components of TF-IDF

TF-IDF consists of two components:

1. **Term Frequency (TF)**
2. **Inverse Document Frequency (IDF)**

---

# 1. Term Frequency (TF)

Term Frequency measures how frequently a term appears in a document.

### Formula

$$
TF(t,d)=\frac{\text{Number of times term } t \text{ appears in document } d}
{\text{Total number of terms in document } d}
$$

Where:

- **t** = Term (word)
- **d** = Document

---

### Example

Suppose a document contains **100 words**, and the word **"cat"** appears **5 times**.

$$
TF(\text{cat})=\frac{5}{100}=0.05
$$

This means **5%** of the words in the document are **"cat"**.

---

# 2. Inverse Document Frequency (IDF)

Inverse Document Frequency measures how unique or rare a word is across the entire corpus.

Words appearing in many documents receive **lower weights**, while rare words receive **higher weights**.

### Formula

$$
IDF(t)=\log\left(\frac{N}{df(t)}\right)
$$

Where:

- **N** = Total number of documents
- **df(t)** = Number of documents containing the term **t**

---

### Interpretation

- If a word appears in **many documents**, its IDF is **small**.
- If a word appears in **very few documents**, its IDF is **large**.

For example:

- **"the"** appears in almost every document → Low IDF
- **"photosynthesis"** appears in only a few documents → High IDF

---

# TF-IDF Score

The TF-IDF score is calculated by multiplying TF and IDF.

### Formula

$$
TF\text{-}IDF(t,d)=TF(t,d)\times IDF(t)
$$

A **high TF-IDF score** indicates that:

- The word appears frequently in the current document.
- The word appears in relatively few other documents.

---

# Example

Suppose we have the following three documents.

### Document 1

```
the cat sat on the mat
```

### Document 2

```
the dog sat on the log
```

### Document 3

```
the cat chased the mouse
```

---

## Step 1: Calculate TF

The word **"cat"** appears **1 time** in Document 1.

Total words in Document 1 = **6**

$$
TF(\text{cat},D_1)=\frac{1}{6}=0.167
$$

---

## Step 2: Calculate IDF

Total documents

$$
N=3
$$

The word **"cat"** appears in:

- Document 1
- Document 3

Therefore,

$$
df(\text{cat})=2
$$

Now,

$$
IDF(\text{cat})=\log\left(\frac{3}{2}\right)
$$

$$
IDF(\text{cat})\approx0.176
$$

---

## Step 3: Calculate TF-IDF

$$
TF\text{-}IDF
=
0.167\times0.176
$$

$$
TF\text{-}IDF
\approx0.029
$$

---

# Example for the Word "the"

The word **"the"** appears in **every document**.

Therefore,

$$
df(\text{the})=3
$$

Hence,

$$
IDF(\text{the})
=
\log\left(\frac{3}{3}\right)
=
\log(1)
=
0
$$

Since

$$
TF\text{-}IDF=TF\times0=0
$$

The word **"the"** receives almost **zero importance**, even though it appears many times.

# Why TF-IDF is Useful?

TF-IDF is widely used in NLP and Information Retrieval for:

- Search Engine Ranking
- Keyword Extraction
- Document Similarity
- Text Classification
- Information Retrieval
- Spam Detection
- News Categorization
- Recommendation Systems
- Document Clustering

---

# Advantages of TF-IDF

### 1. Simple to Understand

- Easy to compute and interpret.

### 2. Computationally Efficient

- Requires very little computation compared to deep learning models.

### 3. Reduces the Importance of Common Words

- Words like **the**, **is**, and **and** receive low scores.

### 4. Highlights Important Keywords

- Gives higher weights to meaningful and informative words.

### 5. Works Well for Traditional Machine Learning

TF-IDF features are commonly used with:

- Naive Bayes
- Logistic Regression
- Support Vector Machines (SVM)
- Random Forest

### 6. Improves Search Relevance

- Search engines use TF-IDF-like scoring to rank documents.

### 7. No Training Required

- Unlike neural embeddings, TF-IDF does not require model training.

---

# Disadvantages of TF-IDF

### 1. Ignores Word Meaning

TF-IDF only counts word frequency.

Example:

```
car
automobile
vehicle
```

Although they have similar meanings, TF-IDF treats them as completely different words.

---

### 2. Ignores Word Order

These two sentences produce similar TF-IDF vectors:

```
Dog bites man
```

```
Man bites dog
```

Even though their meanings are very different.

---

### 3. High-Dimensional Sparse Matrix

Large vocabularies produce thousands of features.

Most values are **0**, leading to sparse matrices.

---

### 4. Cannot Capture Context

TF-IDF does not understand context.

Example:

```
Apple released a new iPhone.
```

```
I ate an apple.
```

The word **Apple** has different meanings, but TF-IDF treats them as the same word.

---

### 5. Sensitive to Vocabulary Size

As the corpus grows, the feature space becomes very large.

---

### 6. Does Not Learn Semantic Relationships

It cannot learn relationships like:

- King → Queen
- Paris → France

Modern embedding models can capture these relationships.

---

### 7. Requires Text Preprocessing

Good TF-IDF performance often depends on preprocessing such as:

- Lowercasing
- Removing stopwords
- Stemming
- Lemmatization
- Tokenization

---

# Intuition Behind TF-IDF

Think of TF-IDF as asking the following question:

> **"Is this word common in this document but uncommon in the rest of the corpus?"**

If the answer is **Yes**, the word is considered important.

For example:

- High frequency in one document → **Important**
- Common across all documents → **Less informative**
- Rare across the corpus but frequent in one document → **Highly informative**

---

# TF-IDF Workflow

```
Raw Text
     │
     ▼
Tokenization
     │
     ▼
Remove Stopwords
     │
     ▼
Stemming/Lemmatization
     │
     ▼
Calculate TF
     │
     ▼
Calculate IDF
     │
     ▼
Multiply TF × IDF
     │
     ▼
TF-IDF Feature Vector
```

---

# Summary

- **TF (Term Frequency)** measures how often a word appears in a document.
- **IDF (Inverse Document Frequency)** measures how rare a word is across the corpus.
- **TF-IDF** is the product of TF and IDF.
- Common words receive **low scores**, while informative and unique words receive **high scores**.
- TF-IDF is simple, fast, and effective for many traditional NLP tasks.
- Modern NLP models often use **word embeddings** (Word2Vec, GloVe, FastText) and **transformer embeddings** (BERT, RoBERTa, etc.) because they capture semantic meaning and context, whereas TF-IDF only measures word importance based on frequency.
