# Basic NLP Terminologies

## 1. Word
A **word** is the smallest meaningful unit of text after tokenization.

**Example:**

Sentence:
```
I love Machine Learning.
```

Words:
```
I, love, Machine, Learning
```

---

## 2. Document
A **document** is a single piece of text. It can be a sentence, paragraph, email, article, review, or any text file.

**Example:**

Document 1:
```
I love Machine Learning.
```

Document 2:
```
Python is easy to learn.
```

Each sentence is considered a separate document.

---

## 3. Corpus
A **corpus** is a collection of multiple documents used for NLP tasks.

**Example:**

Corpus =

Document 1:
```
I love Machine Learning.
```

Document 2:
```
Python is easy to learn.
```

Document 3:
```
NLP is an exciting field.
```

A corpus is simply a dataset of text documents.

---

## 4. Vocabulary
The **vocabulary** is the set of all unique words present in a corpus.

**Example:**

Corpus:

```
I love Python.
I love NLP.
```

Vocabulary:

```
I
love
Python
NLP
```

Vocabulary Size = **4**

> Duplicate words are counted only once.

---

# Tokenization

## What is Tokenization?

**Tokenization** is the process of breaking text into smaller units called **tokens**. These tokens can be words, sentences, or even characters.

It is the **first step** in most NLP pipelines because machine learning models cannot understand raw text directly.

---

## Types of Tokenization

### 1. Word Tokenization

Splits text into individual words.

**Example:**

Input:
```
I love Machine Learning.
```

Output:
```
['I', 'love', 'Machine', 'Learning']
```

---

### 2. Sentence Tokenization

Splits a paragraph into individual sentences.

**Example:**

Input:
```
I love NLP. It is interesting. Python is powerful.
```

Output:
```
[
 'I love NLP.',
 'It is interesting.',
 'Python is powerful.'
]
```

---

## Why is Tokenization Important?

- Converts raw text into machine-readable tokens.
- Forms the foundation of all NLP preprocessing.
- Required before applying techniques like Bag of Words (BoW), TF-IDF, or Word2Vec.
- Makes text easier to analyze and process.

---

## Summary

| Term | Definition |
|-------|------------|
| **Word** | Smallest meaningful unit of text. |
| **Document** | A single text file, sentence, paragraph, or article. |
| **Corpus** | A collection of multiple documents. |
| **Vocabulary** | All unique words present in a corpus. |
| **Tokenization** | Process of splitting text into tokens (words or sentences). |
