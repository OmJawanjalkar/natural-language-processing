# Tokenization - Practical Implementation

Tokenization is the process of splitting text into smaller units called **tokens**. In this notebook, we will use the **NLTK** library to perform both **Sentence Tokenization** and **Word Tokenization**.

```python
!pip install nltk
```

```python
corpus="""Hello Welcome,to Om Jawanjalkar's NLP Tutorials.
Please do watch the entire course! to become expert in NLP.
"""
```

```python
print(corpus)
```

```python
##  Tokenization
## Sentence-->paragraphs
from nltk.tokenize import sent_tokenize

documents=sent_tokenize(corpus)
```

```python
type(documents)
```

```python
for sentences in documents:
    print(sentences)
```

```python
## Tokenization 
## Paragraph-->words
## sentence--->words
from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
```

```python
for sentence in documents:
    print(word_tokenize(sentence))
```

```python
from nltk.tokenize import wordpunct_tokenize
```

```python
wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer
```

```python
tokenizer=TreebankWordTokenizer()
```

```python
tokenizer.tokenize(corpus)
```
