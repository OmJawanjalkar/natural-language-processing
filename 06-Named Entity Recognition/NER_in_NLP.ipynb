# Named Entity Recognition (NER)

## What is Named Entity Recognition (NER)?

**Named Entity Recognition (NER)** is a Natural Language Processing (NLP) technique that identifies and classifies important entities in a text into predefined categories.

These categories include:

- Person
- Organization
- Location
- Date
- Time
- Money
- Percentage
- Product
- Event
- Nationality
- Language

In simple words, NER helps a computer recognize **"who," "where," "when," and "what"** from unstructured text.

---

## Example

**Sentence**

```
Elon Musk is the CEO of Tesla and lives in the United States.
```

**NER Output**

| Word/Phrase | Entity Type |
|-------------|-------------|
| Elon Musk | Person |
| Tesla | Organization |
| United States | Location |

---

## Another Example

**Sentence**

```
Apple launched the iPhone 16 in California on September 9, 2025.
```

**NER Output**

| Entity | Type |
|---------|------|
| Apple | Organization |
| iPhone 16 | Product |
| California | Location |
| September 9, 2025 | Date |

---

# Why Do We Need NER?

NER helps computers understand the meaning of text by identifying important real-world entities.

It is useful for:

- Information Extraction
- Search Engines
- Chatbots
- Question Answering Systems
- News Analysis
- Resume Parsing
- Medical Record Analysis
- Financial Document Analysis

---

# How Does NER Work?

The NER process generally involves the following steps:

```
Raw Text
     │
     ▼
Tokenization
     │
     ▼
Part of Speech (POS) Tagging
     │
     ▼
Named Entity Recognition
     │
     ▼
Extracted Entities
```

For example,

```
Sentence

Barack Obama visited India in 2015.
```

After NER,

```
Barack Obama → Person
India → Location
2015 → Date
```

| Entity Type | Description | Example |
|--------------|-------------|---------|

---

# Limitations

- Ambiguous names can be difficult to classify.

Example:

```
Apple released a new MacBook.
```

Here,

```
Apple → Organization
```

But,

```
I ate an apple.
```

Here,

```
apple → Fruit (not an organization)
```

The meaning depends on the context.

---

# NER in NLTK

In NLTK, Named Entity Recognition is performed after:

1. Tokenization
2. Part of Speech (POS) Tagging
3. Named Entity Chunking

Pipeline:

```
Sentence
   │
   ▼
Word Tokenization
   │
   ▼
POS Tagging
   │
   ▼
Named Entity Chunking
   │
   ▼
Recognized Entities
```
