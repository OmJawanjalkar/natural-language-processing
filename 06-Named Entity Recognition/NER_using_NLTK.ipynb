# Named Entity Recognition (NER)

Named Entity Recognition (NER) is an NLP technique that identifies and classifies named entities in text into predefined categories such as Person, Organization, Location, Date, Time, Money, Product, and Event. It helps machines extract structured information from unstructured text and is widely used in search engines, chatbots, information extraction, and question-answering systems.

---

```python
import nltk
from nltk.corpus import stopwords
from nltk import  pos_tag, ne_chunk, word_tokenize, sent_tokenize
```

```python
nltk.download("punkt")
nltk.download("punkt_tab")
nltk.download("averaged_perceptron_tagger")
nltk.download("averaged_perceptron_tagger_eng")
nltk.download("maxent_ne_chunker")
nltk.download("maxent_ne_chunker_tab")
nltk.download("words")
```

```python
text = """
Elon Musk is the CEO of Tesla.
He lives in the United States.
Microsoft was founded by Bill Gates.
"""
```

```python
words = word_tokenize(text)
print("Words:", words)
```

```python
pos_tags = pos_tag(words)
print("POS Tags:", pos_tags)
```

```python
named_entities = ne_chunk(pos_tags)
print("Named Entities:", named_entities)
```

```python
for subtree in named_entities:
    if hasattr(subtree, "label"):
        print(subtree.label(), ":", " ".join(token for token, pos in subtree.leaves()))
```



---


| Function          | Purpose                                                   |
| ----------------- | --------------------------------------------------------- |
| `word_tokenize()` | Splits text into words                                    |
| `pos_tag()`       | Assigns Part-of-Speech tags                               |
| `ne_chunk()`      | Identifies named entities                                 |
| `draw()`          | Displays the NER tree graphically                         |
| `label()`         | Returns the entity type (PERSON, ORGANIZATION, GPE, etc.) |


### Common NER Labels

| Label        | Meaning                    |
| ------------ | -------------------------- |
| PERSON       | Name of a person           |
| ORGANIZATION | Company or institution     |
| GPE          | Country, State, City       |
| LOCATION     | Physical location          |
| FACILITY     | Building, Airport, Highway |
| GSP          | Geographical entity        |


#### Summary

    word_tokenize() → Splits text into words.
    pos_tag() → Assigns grammatical tags to each word.
    ne_chunk() → Recognizes and classifies named entities.
    The output can identify entities such as Person, Organization, and Location, making it useful for information extraction and many other NLP applications.
